##### indexing and slicing with reindex() and set_index()

```python
# ******************************************************************************************
# So Reindex is frequently used when working with Time series data where we have a so called datetime index, 
# but we can also use Reindex for other purposes.
# And one advantage of the Reindex method is that we can pass row labels or column labels to Reindex that are not in the data frame.
# ******************************************************************************************
# As a side note, we always have to pass here a list to the index and columns parameter, even if we only have one element.
```

```python
file_path_summer = r"C:\Users\aitor\OneDrive\Desktop\FRED\course_env\Pandas_2025\onthego_materials\Course_Materials_Part1\Video_Lecture_NBs\summer.csv"
summer = pd.read_csv(file_path_summer)
summer
```

```python
summer.loc[[0,5,30000],[ "Athlete", "Medal"]]
```

```python
# KeyError: '[40000] not in index'
summer.loc[[0,5,30000, 40000],[ "Athlete", "Medal"]]
```

```python
# And actually for the row index label 40,000, we get an additional row here with missing values in the columns athlete and medal.
# and for the column age, we get a new column with missing values in the rows 0, 5, and 30,000.
summer.reindex(index=[0,5,30000, 40000], columns=["Athlete", "Medal", "Age"])
```

```python
# reindex limitation
# ******************************************************************************************
# And actually, the Reindex method does not work in case we do not have a unique column index or row
# ******************************************************************************************
# it will work for columns but generate an error for rows
```

```python
summer = pd.read_csv(file_path_summer, index_col='Athlete')
summer
```

```python
# it will work for columns but generate an error for rows
summer.reindex(columns=["Medal", "Age"])
```

```python
# ValueError: cannot reindex on an axis with duplicate labels
summer.reindex(index = ["PHELPS, Michael"],columns=["Medal", "Age"])
```
