#### Porcentages para contrapartida y banca y variaciones entre m y m1

```python
# paso 1 - BASE de calculos: filtramos las columnas que nos interesan de la comparativa de conteo de contratos y sumatorio de rwas
# --------------------------------

comparison_renamed_filtered = comparison_renamed.loc[:,['banca', 'ctpa','rwas_Tot_m','rwas_Tot_m1','rwas_Tot_diff']]
comparison_renamed_filtered
```

```python
# === 2. Porcentajes por banca y ctpa ===

comparison_renamed_filtered['pct_var_rwas_MoM'] = (
    ((comparison_renamed_filtered['rwas_Tot_m'] - comparison_renamed_filtered['rwas_Tot_m1']) / 
     comparison_renamed_filtered['rwas_Tot_m1'].replace(0, pd.NA))
)

grand_total_m = comparison_renamed_filtered.loc[comparison_renamed_filtered['banca'] == 'Total', 'rwas_Tot_m'].values[0]
grand_total_m1 = comparison_renamed_filtered.loc[comparison_renamed_filtered['banca'] == 'Total', 'rwas_Tot_m1'].values[0]

comparison_renamed_filtered['Grand_Tot_m'] = grand_total_m
comparison_renamed_filtered['Grand_Tot_m1'] = grand_total_m1

# Compute the new column
comparison_renamed_filtered['pct_Grand_Tot_m'] = comparison_renamed_filtered['rwas_Tot_m'] / grand_total_m
comparison_renamed_filtered['pct_Grand_Tot_m1'] = comparison_renamed_filtered['rwas_Tot_m1'] / grand_total_m1

comparison_renamed_filtered['pct_var_GT_MoM'] = (
    ((comparison_renamed_filtered['pct_Grand_Tot_m'] - comparison_renamed_filtered['pct_Grand_Tot_m1']) / 
     comparison_renamed_filtered['pct_Grand_Tot_m1'].replace(0, pd.NA))
)

# Totales por banca y ctpa
comparison_renamed_filtered['ctpa_Tot_m'] = comparison_renamed_filtered.groupby(['ctpa'])['rwas_Tot_m'].transform('sum')
comparison_renamed_filtered['ctpa_Tot_m1'] = comparison_renamed_filtered.groupby(['ctpa'])['rwas_Tot_m1'].transform('sum')

comparison_renamed_filtered['ctpa_Tot_diff'] = comparison_renamed_filtered['ctpa_Tot_m'] - comparison_renamed_filtered['ctpa_Tot_m1']

# % por ctpa dentro de banca (m y m1)
comparison_renamed_filtered['pct_ctpa_Tot_m'] = (comparison_renamed_filtered['rwas_Tot_m'] / comparison_renamed_filtered['ctpa_Tot_m'])
comparison_renamed_filtered['pct_ctpa_Tot_m1'] = (comparison_renamed_filtered['rwas_Tot_m1'] / comparison_renamed_filtered['ctpa_Tot_m1'])

comparison_renamed_filtered['pct_var_ctpa_Tot_MoM'] = (
    ((comparison_renamed_filtered['pct_ctpa_Tot_m'] - comparison_renamed_filtered['pct_ctpa_Tot_m1']) / 
     comparison_renamed_filtered['pct_ctpa_Tot_m1'].replace(0, pd.NA))
)

ordered_list = [
    'banca',
    'ctpa',
    'rwas_Tot_m',
    'Grand_Tot_m',
    'pct_Grand_Tot_m',
    'rwas_Tot_m1',
    'Grand_Tot_m1',
    'pct_Grand_Tot_m1',
    'rwas_Tot_diff',
    'pct_var_rwas_MoM',
    'pct_var_GT_MoM',
    'ctpa_Tot_m',
    'pct_ctpa_Tot_m',
    'ctpa_Tot_m1',
    'pct_ctpa_Tot_m1',
    'ctpa_Tot_diff',
    'pct_var_ctpa_Tot_MoM'
]

comparison_renamed_filtered_final = comparison_renamed_filtered.loc[:,ordered_list]

cols_to_blank_zero = [
    'ctpa_Tot_m',
    'ctpa_Tot_m1',
    'pct_ctpa_Tot_m',
    'pct_ctpa_Tot_m1',
    'ctpa_Tot_diff',
    'pct_var_ctpa_Tot_MoM'
    
]

# Set values in last row (index 12) to blank (None or NaN)
#df.loc[12, cols_to_blank] = None
#comparison_renamed_filtered_final.loc[comparison_renamed_filtered_final['area_id'] == 'Total', cols_to_blank_zero] = None

# Set those columns to 0.00 for the 'Total' row (row 12 or where area_id == 'Total')
comparison_renamed_filtered_final.loc[comparison_renamed_filtered_final['banca'] == 'Total', cols_to_blank_zero] = 0.0000


#comparison_renamed_filtered_final.loc['12',['ctpa_total_m','ctpa_total_m1', 'pct_ctpa_totales_m', 'pct_ctpa_totales_m1','pct_ctpa_mom_totales_rel']] = 0.0000
```

```python
| **Column Name**            | **Description**                                                                                                                                                  |
| -------------------------- | ---------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| `area_id`                  | The identifier for a geographic or organizational **area**. Example: `"0023"`, `"0058"`, etc.                                                                    |
| `ctpa_id`                  | The identifier for a specific **CTPA** (could refer to a cost type, product, service line, or account category).                                                 |
| `rwas_Total_m`             | Total APRs de la contrapartida en la banca asociada en el mes actual (m). (ya es un agrupado que representa el 100% de los APRs de la contrapartida en la banca asociada).                             |
| `rwas_Total_m1`            | Total APRs de la contrapartida en la banca asociada en el mes anterior (m1). (ya es un agrupado que representa el 100% de los APRs de la contrapartida en la banca asociada).                                                                                         |
| `rwas_Total_diff`          | Variacion neta de total de APRs de la contrapartida en la banca asociada en el mes actual (m).vs mes anterior (m1). `rwas_Total_m - rwas_Total_m1`.  
|
| `pct_ctpa_mom_rel`         | Variacion relativa de total de APRs de la contrapartida en la banca asociada en el mes actual (m).vs mes anterior (m1). `rwas_Total_diff / rwas_Total_m1`. Expresses growth or decline as a proportion.                               |
                     
| `ctpa_total_m`             | Total neto de APRs que representa esa contrapartida en el mes actual respecto al total de APRs de la misma contrapartida.                                                                 |
| `ctpa_total_m1`            | Total neto de APRs que representa esa contrapartida en el mes anterior respecto al total de APRs de la misma contrapartida.
                     
| `pct_ctpa_totales_m`       | Peso en porcentaje de APRs que representa esa contrapartida en el mes actual respecto al total de APRs de la misma contrapartida.(`rwas_Total_m / ctpa_total_m`.)                      |
| `pct_ctpa_totales_m1`      | Peso en porcentaje de APRs que representa esa contrapartida en el mes anterior respecto al total de APRs de la misma contrapartida. (`rwas_Total_m1 / ctpa_total_m1`.)                  |
| `pct_ctpa_mom_totales_rel` |Variacion relativa del peso en porcentaje de APRs que representa esa contrapartida en el mes actual respecto al peso en porcentaje de APRs de la misma contrapartida en el mes anterior. (`pct_ctpa_totales_m / pct_ctpa_totales_m1 - 1`, showing how the proportion of contribution changed. )
```
